```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = "/content/drive/MyDrive/8th_SEM_MP_2/DDI_types.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Set figure size
plt.figure(figsize=(12, 6))

# Create a bar plot
sns.countplot(data=df, x="merged DDI type index", palette="viridis")

# Add labels and title
plt.xlabel("Merged DDI Type Index")
plt.ylabel("Count")
plt.title("Distribution of Merged DDI Type Indices")

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the occurrences of each DDI type
ddi_counts = df["merged DDI type index"].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 8))
ddi_counts.plot.pie(autopct="%1.1f%%", cmap="Set3", startangle=90)

# Add title
plt.title("Proportion of Origin DDI Types")

# Show the plot
plt.ylabel("")  # Remove y-label
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all text from the "Origin DDI's type" column
text = " ".join(df["Origin DDI's type"].astype(str))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

# Display the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of DDI Types")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["merged DDI type index"], color="lightblue")

plt.title("Box Plot of Merged DDI Type Indices")
plt.ylabel("Merged DDI Type Index")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["merged DDI type index"], bins=20, kde=True, color="green")

plt.title("Histogram of Merged DDI Type Indices")
plt.xlabel("Merged DDI Type Index")
plt.ylabel("Frequency")

plt.show()
```

Violin Plot: Distribution of Merged DDI Type Indices
Violin plots show data distribution with density estimation.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(y=df["merged DDI type index"], color="purple")

plt.title("Violin Plot of Merged DDI Type Indices")
plt.ylabel("Merged DDI Type Index")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np # Import numpy for array operations

# Sample data (replace with your actual data)
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [4, 6, 2, 8, 5]

# Number of variables
num_vars = len(categories)

# Angle for each axis
angles = [n / float(num_vars) * 2 * 3.14159 for n in range(num_vars)]
angles += angles[:1]

# Initialize the spider plot
ax = plt.subplot(111, polar=True)

# Plot the data
ax.plot(angles, values + values[:1], 'o-', linewidth=2)

# Fill the area
ax.fill(angles, values + values[:1], alpha=0.25)

# Set the labels for each axis
# Convert angles[:-1] to a NumPy array for element-wise operations
ax.set_thetagrids(np.array(angles[:-1]) * 180/3.14159, categories)

# Set the limits for the radial axis
plt.ylim(0, max(values) * 1.2) # Adjust the limit as needed

# Add a title
plt.title("Spider Plot")

# Display the plot
plt.show()
```
